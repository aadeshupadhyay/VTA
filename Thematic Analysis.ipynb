```python
import pandas as pd

lecture=pd.read_csv("human_evaluation/evaluated_rajashree/lecture.csv")
guidance=pd.read_csv("human_evaluation/evaluated_rajashree/homework_guidance.csv")
clarification=pd.read_csv("human_evaluation/evaluated_rajashree/homework_clarification.csv")
```

```python
filter_df=pd.DataFrame()
selected_score=["Moderately Relevant","Perfectly Relevant","Highly Relevant"]
filtered_df = lecture[lecture['evaluation_gemma'].isin(selected_score) & ~lecture['evaluation_llama'].isin(selected_score)]
```

```python
filtered_guidance = guidance[guidance['evaluation_gemma'].isin(selected_score) & ~guidance['evaluation_llama'].isin(selected_score)]
filtered_clarification=clarification[clarification['evaluation_gemma'].isin(selected_score) & ~clarification['evaluation_llama'].isin(selected_score)]
```

```python
filtered_df.to_csv("thematic_analysis/lecture_thematic.csv")
filtered_guidance.to_csv("thematic_analysis/guidance_thematic.csv")
filtered_clarification.to_csv("thematic_analysis/clarification_thematic.csv")
```
